```python
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
# to get amount/percentage of missing Data
null_var=df.isnull().sum()/df.shape[0]*100
null_var
```

```python
# columns who's missing data percent is greater than 17
drop_columns=null_var[null_var>17].keys()
drop_columns
```

```python
df2_drop_clm=df.drop(columns=drop_columns)
```

```python
df2_drop_clm.shape
```

```python
sns.heatmap(df2_drop_clm.isnull())
```

```python
# delete the rows who have value NAN

#like if any of the row have single NAN value it will delete that particular row 
df3_drop_rows=df2_drop_clm.dropna()
```
